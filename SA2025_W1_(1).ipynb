### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# Calculate median mpg
mpg_median = df['mpg'].median()

# Filter cars with mpg above median
above_median_mpg = df[df['mpg'] > mpg_median]

# Find max hp_per_weight among filtered cars (index is hp_per_weight)
max_ratio = above_median_mpg.index.max()

# Get all cars with that max ratio
best_cars = above_median_mpg.loc[above_median_mpg.index == max_ratio]

print(best_cars)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Reset index if needed (here index is hp_per_weight, we don't need it)
df_reset = df.reset_index()

# Calculate average mpg by model_year and origin
avg_mpg = df_reset.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Plot with seaborn lineplot
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Reset index to avoid any issues with index being hp_per_weight
df_reset = df.reset_index()

plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=df_reset,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(20, 200),  # adjust point sizes for visibility
    hue_order=['japan', 'europe', 'usa'],
    palette='deep',
    alpha=0.7,
    edgecolor='w',
    linewidth=0.7
)

plt.title('Horsepower vs Weight Colored by Origin and Sized by MPG')
plt.xlabel('Horsepower')
plt.ylabel('Weight')

# Improve legend handling for size and hue
handles, labels = scatter.get_legend_handles_labels()
# The legend includes size and hue, so split them:
n_hue = len(df_reset['origin'].unique())
scatter.legend(handles=handles[:n_hue+1] + handles[n_hue+2:], labels=labels[:n_hue+1] + labels[n_hue+2:], title='Origin & MPG')

plt.grid(True)
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# Group by horsepower, weight, origin as a proxy for car model
group_cols = ['horsepower', 'weight', 'origin']

# Count unique model_year appearances per group
year_counts = df.groupby(group_cols)['model_year'].nunique()

# Select groups appearing in multiple years (≥ 2)
multi_year_groups = year_counts[year_counts >= 2].index

# Filter df for those groups
df_multi_year = df.set_index(group_cols).loc[multi_year_groups].reset_index()

# Compute mpg std dev, count appearances, and average mpg
stats = df_multi_year.groupby(group_cols).agg(
    mpg_std=('mpg', 'std'),
    appearances=('model_year', 'count'),
    avg_mpg=('mpg', 'mean')
).reset_index()

# Filter for low variation (std < 1.0)
consistent_groups = stats[stats['mpg_std'] < 1.0]

# Sort by appearances desc, then avg_mpg desc
consistent_sorted = consistent_groups.sort_values(
    by=['appearances', 'avg_mpg'],
    ascending=[False, False]
)

print(consistent_sorted)
```
